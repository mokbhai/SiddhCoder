```python
txt="Hello All. We're hoping you guys are doing great."

# this gives number of sentences
print(len(txt.split('.')))

# this gives number of words
print(len(txt.split()))
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize

sentences = sent_tokenize(txt)
print(len(sentences))
words = word_tokenize(txt)
print(len(words))
```

```python
print(words)

# Formating the words
print("\nFormated Words: ")
for word in words:
    if (word != '.'):
        print(word)
```
